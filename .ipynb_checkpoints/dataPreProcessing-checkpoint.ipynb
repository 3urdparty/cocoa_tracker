```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


def clean_price_column(dataframe, columns):
    """
    Cleans specified columns by removing dollar signs and commas,
    and converts the values to numeric format.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data.
        columns (list): List of column names to clean.

    Returns:
        pd.DataFrame: The modified DataFrame.
    """
    for column in columns:
        dataframe[column] = dataframe[column].str.replace('$', '').str.replace(',', '')
        dataframe[column] = pd.to_numeric(dataframe[column])
    return dataframe


def drop_unnecessary_columns(dataframe, columns):
    """
    Drops specified columns from the DataFrame.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data.
        columns (list): List of column names to drop.

    Returns:
        pd.DataFrame: The modified DataFrame.
    """
    return dataframe.drop(columns=columns)


def remove_outliers(dataframe, quantile_range=(0.10, 0.90)):
    """
    Removes outliers by trimming the lowest and highest quantiles.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data.
        quantile_range (tuple): The lower and upper quantile thresholds.

    Returns:
        pd.DataFrame: The modified DataFrame.
    """
    numeric_cols = dataframe.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        lower_quantile = dataframe[col].quantile(quantile_range[0])
        upper_quantile = dataframe[col].quantile(quantile_range[1])
        dataframe = dataframe[(dataframe[col] >= lower_quantile) & (dataframe[col] <= upper_quantile)]
    return dataframe


def normalize_numeric_columns(dataframe):
    """
    Normalizes numeric columns using StandardScaler.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data.

    Returns:
        pd.DataFrame: The modified DataFrame with normalized numeric columns.
    """
    numeric_cols = dataframe.select_dtypes(include=['float64', 'int64']).columns
    scaler = StandardScaler()
    dataframe[numeric_cols] = scaler.fit_transform(dataframe[numeric_cols])
    return dataframe


def encode_categorical_columns(dataframe, exclude_columns):
    """
    Creates dummy variables for categorical columns, excluding specified columns.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data.
        exclude_columns (list): List of column names to exclude from encoding.

    Returns:
        pd.DataFrame: The modified DataFrame with encoded categorical columns.
    """
    if exclude_columns is None:
        exclude_columns = []
    categorical_cols = dataframe.select_dtypes(include=['object']).columns.difference(exclude_columns)
    dataframe = pd.get_dummies(dataframe, columns=categorical_cols, drop_first=True, dtype=int)
    return dataframe


def main():
    data = pd.read_csv('Airbnb_Open_Data.csv', low_memory=False)

    data = clean_price_column(data, ['price', 'service fee'])

    columns_to_drop = ['license', 'house_rules', 'lat', 'long', 'country', 'NAME', 'host name', 'host id',
                       'country code', 'last review', 'id']
    data = drop_unnecessary_columns(data, columns_to_drop)

    data = data.dropna().drop_duplicates()

    data = remove_outliers(data)

    data = normalize_numeric_columns(data)

    data = encode_categorical_columns(data, exclude_columns=['last review', 'neighbourhood'])
    le = LabelEncoder()
    data['neighbourhood'] = le.fit_transform(data['neighbourhood'])
    return data  # Return the processed data


if __name__ == "__main__":
    processed_data = main()  # Assign returned data to a variable that is accessible globally
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming processed_data is your DataFrame

# Select only the numeric columns
numeric_data = processed_data.select_dtypes(include='number')

# Calculate the correlation matrix for numeric columns
corr_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap (Numeric Columns)')

# Rotate x-axis labels
plt.xticks(rotation=90)

plt.show()
processed_data
```

```python
# import pydot

target = np.array(processed_data['price'])
rfData = processed_data.drop('price', axis=1)
rfDataList = list(rfData.columns)
rfData = np.array(rfData)
rfData

```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

train_rf, test_rf, train_target, test_target = train_test_split(rfData, target, test_size=.25, random_state=42)
print('trainging features: ', train_rf.shape)
print('trainging labels: ', train_target.shape)
print('trainging features: ', train_target.shape)
print('trainging features: ', test_target.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(train_rf, train_target)
predictions = model.predict(test_rf)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_target, predictions)
print(mse)
```
